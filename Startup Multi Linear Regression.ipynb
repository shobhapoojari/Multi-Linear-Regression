```python
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
```

```python
startnew.corr()
```

# Creating new columns which are transformations for the existing columns

```python
startnew['admin_sq']=startnew.Administration**2
startnew['admin_sqrt']=startnew.Administration**0.5
startnew['admin_log']=np.log(startnew.Administration)
startnew['mar_sq']=startnew.Marketing_Spend**2
startnew['mar_sqrt']=startnew.Marketing_Spend**0.5
startnew['mar_log']=np.log(startnew.Marketing_Spend)
startnew['rnd_sq']=startnew.RnD_Spend**2
startnew['rnd_sqrt']=startnew.RnD_Spend**0.5
startnew['rnd_log']=np.log(startnew.RnD_Spend) 
```

```python
startnew
```

# Model Creation

```python
model=smf.ols('Profit~RnD_Spend+Administration+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
model.params
```

```python
model.fittedvalues
```

```python
model.rsquared
```

```python
model.summary()
```

```python
#t and p-Valuesvalue
print(model.tvalues, '\n', model.pvalues)
```

```python
sns.boxplot(model.resid)
```

```python
sns.distplot(model.resid)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

# Normal Q-Q Plot of Residuals

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

# Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
model.pvalues
```

# Calculation VIF

```python
rsd_admin = smf.ols('Administration~RnD_Spend+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit().rsquared
vif_admin = 1/(1-rsd_admin)

rsd_rnd = smf.ols('RnD_Spend~Administration+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit().rsquared
vif_rnd = 1/(1-rsd_rnd)

rsq_mar = smf.ols('Marketing_Spend~RnD_Spend+Administration+State_California+State_Florida+State_NewYork',data=startnew).fit().rsquared
vif_mar = 1/(1-rsq_mar)


#Storing vif value in a data frame
d1 = {'Variables':['admin','rnd','mar'],'VIF':[vif_admin,vif_rnd,vif_mar]}
vif_frame = pd.DataFrame(d1)
vif_frame
```

# Administration and Marketing spend has high pvalues particularly administration which has a pvalue of 60%

# Calculating cook's distance and checking for values above the leverge cutoff

```python
model_influence = model.get_influence()
(c,_) = model_influence.cooks_distance
k=6
n=50
leverage_cutoff = 3*((k + 1)/n)
leverage_cutoff
for i in c:
    if i>leverage_cutoff:
        print(i)
print(c)
```

# All values are below the leverage cutoff

# Creting models with different with different transformed columns instead of the Administration column

```python
model_admin_sq=smf.ols('Profit~RnD_Spend+admin_sq+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
model_admin_sqrt=smf.ols('Profit~RnD_Spend+admin_sqrt+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
model_admin_log=smf.ols('Profit~RnD_Spend+admin_log+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
```

```python
print(model_admin_sq.pvalues,model_admin_sqrt.pvalues,model_admin_log.pvalues,sep='\n')
```

# Even with transformations applied, the pvalues is very high. Thus eliminating the Administrations column

```python
model1=smf.ols('Profit~RnD_Spend+Marketing_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
model_mar_sq=smf.ols('Profit~RnD_Spend+mar_sq+State_California+State_Florida+State_NewYork',data=startnew).fit()
model_mar_sqrt=smf.ols('Profit~RnD_Spend+mar_sqrt+State_California+State_Florida+State_NewYork',data=startnew).fit()
print(model_mar_sq.pvalues,model_mar_sqrt.pvalues,model1.pvalues,sep='\n')
```

# P-values of marketing spend is higher than 5% in all the transformation and in the untransformed data.Thus marketing spend column is eliminated

# Calculating cook's distance and checking for values above the leverage cutoff

```python
model_influence = model1.get_influence()
(c, _) = model_influence.cooks_distance
k=5
n=50
leverage_cutoff = 3*((k + 1)/n)
leverage_cutoff
for i in c:
    if i>leverage_cutoff:
        print(i)
print(c)    
print(leverage_cutoff)
```

# All values are below the leverge cutoff

# Calculating cook's distance and checking for values above the leverage cutoff

```python
model_influence = model2.get_influence()
(c, _) = model_influence.cooks_distance
k=4
n=50
leverage_cutoff = 3*((k + 1)/n)
leverage_cutoff
startnew['c']=c
startnew=startnew[startnew.c<leverage_cutoff] 
startnew
```

# Removed data that was above the leverage cutoff

```python
model3=smf.ols('Profit~RnD_Spend+State_California+State_Florida+State_NewYork',data=startnew).fit()
print(model3.rsquared,model2.rsquared)
```

```python
print(np.sqrt(model3.mse_resid),np.sqrt(model2.mse_resid))
```
